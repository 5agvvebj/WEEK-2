````python
# Import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
from glob import glob
import warnings

warnings.filterwarnings('ignore')

# Define paths
train_path = r"F:\Grisha Data\grisha college\aicte greenai internship +project\archive\DATASET\TRAIN"
test_path = r"F:\Grisha Data\grisha college\aicte greenai internship +project\archive\DATASET\TEST"

# Initialize empty lists to store image data and labels
x_data = []
y_data = []

# Iterate through each category folder in the training dataset
for category in glob(train_path + '/*'):
    # Loop through each image in the category folder
    for file in tqdm(glob(category + '/*'), desc=f"Processing {os.path.basename(category)}"):
        img_array = cv2.imread(file)  # Read the image
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        x_data.append(img_array)  # Append image data
        y_data.append(os.path.basename(category))  # Append category label

# Create a DataFrame with the image data and corresponding labels
data = pd.DataFrame({'image': x_data, 'label': y_data})

# Check the first few rows of the DataFrame
print(data.head())

# Visualization: Pie chart to show label distribution
colors = ['#a0d157', '#c48bb8']  # Custom colors for the pie chart
labels = data['label'].unique()  # Dynamically extract unique labels

plt.pie(
    data['label'].value_counts(),
    labels=labels,  # Use dynamic labels
    autopct='%0.2f%%',  # Show percentage values
    colors=colors,
    startangle=90,
    explode=[0.05] * len(labels)  # Add a slight separation for each slice
)
plt.title("Label Distribution")
plt.show()
````

Sure! Let's break the code down step by step in simple terms so even a beginner can understand:

---

### **What Does the Code Do?**
This code processes images stored in folders, assigns them labels based on their folder names, and then visualizes how many images belong to each label (category) using a pie chart.

---

### **Step-by-Step Explanation:**

#### **1. Importing Libraries**
```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
from glob import glob

warnings.filterwarnings('ignore')
```
- These libraries are tools to help us with:
  - **os**: Handle file and folder paths.

  - **pandas (pd)**: Manage data in tables.
  - **matplotlib.pyplot (plt)**: Create graphs and visualizations.
  - **cv2**: Read and process image files.
  - **tqdm**: Show a progress bar when processing lots of files.
  - **glob**: Find files in folders using patterns.
  - **warnings**: Suppress warnings that might pop up unnecessarily.

#### **2. Define File Paths**
```python
train_path = r"F:\Grisha Data\grisha college\aicte greenai internship +project\archive\DATASET\TRAIN"
```

- **What is this?**

#### **3. Initialize Lists to Store Data**

---

#### **4. Read Images and Assign Labels**
```python
for category in glob(train_path + '/*'):
    for file in tqdm(glob(category + '/*'), desc=f"Processing {os.path.basename(category)}"):
        img_array = cv2.imread(file)  # Read the image
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        x_data.append(img_array)  # Add the image to the list
        y_data.append(os.path.basename(category))  # Add the category name
```
- **What's Happening Here?**
  - **Outer Loop**: Goes through each folder (category) in `train_path`.
  - **Inner Loop**: Reads every image file in the folder.
  - **cv2.imread(file)**: Reads the image into a format your program can understand.
  - **cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)**: Converts the image's color format from BGR (used by OpenCV) to RGB (used by most other tools).
  - `x_data.append(img_array)`: Stores the image data in the list.
  - `y_data.append(os.path.basename(category))`: Extracts the folder name (category) and stores it as the label.

**Example**:
- If the folder name is "Organic" and it contains an image called "img1.jpg":
  - `x_data` will store the image data of "img1.jpg".
  - `y_data` will store the label "Organic".

---

```python

```
- **What is this?**

    - Column 2 (`label`) contains the category of each image.

# CNN- CONVOLUTION NEURAL NETWORK

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Activation, MaxPooling2D, Flatten, Dense, Dropout
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(128, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(2))
model.add(Activation('sigmoid'))

model.compile(loss = "binary_crossentropy",
              optimizer = "adam",
              metrics = ["accuracy"])
batch_size = 256
````

````python
model.summary()
````

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255)
````

````python
test_datagen =  ImageDataGenerator(rescale=1./255)
````

````python
train_generator = train_datagen.flow_from_directory(
    train_path, 
    target_size = (224, 224),
    batch_size = batch_size, 
    color_mode = "rgb", 
    class_mode = "categorical")

test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size = (224, 224),
    batch_size = batch_size,
    color_mode = "rgb",
    class_mode = "categorical")
````

````python
hist = model.fit(
    train_generator, 
    epochs=10, 
    validation_data=test_generator
)
````
